```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]

df = pd.DataFrame(prices,
                  columns=["Month","Price"]          
                  )
df.set_index("Month", inplace=True)
priceUSD = [i[1]/4 for i in prices]
df['PriceUSD'] = priceUSD
df['PriceUSD'].plot(linestyle='dashed', color='red')
plt.title("Price of goods (USD)")
plt.grid()
plt.show()
```
